```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set a style for clean, professional visuals
sns.set_theme(style="whitegrid")

# 1. Generate Mock Raw Data
np.random.seed(42)
dates = pd.date_range(start="2026-01-01", periods=100, freq="D")
categories = np.random.choice(["Electronics", "Clothing", "Home & Kitchen", "Books"], size=100)
revenue = np.random.normal(loc=500, scale=150, size=100) + (np.arange(100) * 2)  # upward trend
marketing_spend = revenue * 0.15 + np.random.normal(loc=10, scale=15, size=100)

# Create DataFrame
df = pd.DataFrame({
    "Date": dates,
    "Category": categories,
    "Revenue": revenue,
    "Marketing_Spend": marketing_spend
})

print("Raw Data Sample:")
print(df.head())
print("-" * 50)
```

```python
# --- CHART 1: LINE CHART ---
plt.figure(figsize=(10, 5))

# Plotting the trend
sns.lineplot(data=df, x="Date", y="Revenue", color="#1f77b4", linewidth=2.5)

# Crafting the story & design
plt.title("Daily Revenue Growth Trend (Q1 2026)", fontsize=14, fontweight='bold', pad=15)
plt.xlabel("Date", fontsize=12)
plt.ylabel("Revenue ($)", fontsize=12)
plt.tight_layout()

# To show in VS Code
plt.show()
```

```python
# --- CHART 2: BAR CHART ---
plt.figure(figsize=(8, 5))

# Aggregate data first for a cleaner bar chart
category_sales = df.groupby("Category")["Revenue"].sum().reset_index().sort_values(by="Revenue", ascending=False)

# Plotting with a beautiful palette
sns.barplot(data=category_sales, x="Category", y="Revenue", palette="Blues_d")

# Enhancing understanding
plt.title("Total Revenue Contribution by Product Category", fontsize=14, fontweight='bold', pad=15)
plt.xlabel("Product Category", fontsize=12)
plt.ylabel("Total Revenue ($)", fontsize=12)
plt.tight_layout()

plt.show()
```

```python
# --- CHART 3: SCATTER PLOT ---
plt.figure(figsize=(8, 6))

# Plotting with a trend line (regression line) to clearly reveal insights
sns.regplot(data=df, x="Marketing_Spend", y="Revenue", 
            scatter_kws={"alpha":0.7, "color":"#070907"}, line_kws={"color":"red"})

# Designing for decision-making
plt.title("Impact of Marketing Spend on Revenue", fontsize=14, fontweight='bold', pad=15)
plt.xlabel("Marketing Spend ($)", fontsize=12)
plt.ylabel("Revenue ($)", fontsize=12)

# Adding a helpful note directly on the chart
plt.text(15, 800, "Strong positive correlation:\nMore spend = More revenue", 
         fontsize=10, bbox=dict(facecolor='white', alpha=0.8))

plt.tight_layout()
plt.show()
```

```python
# --- CHART 4: BOX PLOT ---
plt.figure(figsize=(9, 6))

# Plotting distribution
sns.boxplot(data=df, x="Category", y="Revenue", palette="Set2")

# Design adjustments
plt.title("Revenue Distribution & Volatility by Category", fontsize=14, fontweight='bold', pad=15)
plt.xlabel("Category", fontsize=12)
plt.ylabel("Transaction Revenue ($)", fontsize=12)
plt.tight_layout()

plt.show()
```
